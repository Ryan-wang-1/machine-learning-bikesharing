```python
# 导入必要的工具包
# 数据读取及基本处理
import numpy as np
import pandas as pd
#模型
from sklearn.model_selection import GridSearchCV      #用GridSearchCV做
#模型评估  
from sklearn.metrics import  log_loss        #评价指标
#图形可视化
import matplotlib.pyplot as plt
```

```python
#读入数据
train = pd.read_csv("FE_pima-indians-diabetes.csv")
print("Shape of dataset::{}".format(train.shape))
train.head()
#print("train : " + str(train.shape))
```

```python
train.info()
```

## 准备训练数据

```python
# get labels
y_train = train['Target']   
X_train = train.drop(['Target'],axis=1)

#用于特征重要性可视化
feat_names=X_train.columns
```

```python
# 用交叉验证进行模型选择
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.model_selection import cross_val_score
loss=cross_val_score(lr,X_train,y_train,cv=5,scoring='neg_log_loss')
print('logloss',-loss)
print('cv.logloss is:',-loss.mean())
```

## L2正则的默认参数：0.4761

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
penaltys=['l1','l2']
cs=[0.001,0.01,0.1,1,10,100,1000]
turned_parameters=dict(penalty=penaltys,C=cs)
lr_penalty= LogisticRegression()
grid=GridSearchCV(lr_penalty,turned_parameters,cv=5,scoring='neg_log_loss',n_jobs=1,verbose=5)
grid.fit(X_train,y_train)
```

```python
print(-grid.best_score_)
print(grid.best_params_)
```

## LI正则c=1，0.4760，相差不大

```python
penaltys=['l1','l2']
cs=[0.001,0.01,0.1,1,10,100,1000]
turned_parameters=dict(penalty=penaltys,C=cs)
lr_penalty= LogisticRegression()
grid=GridSearchCV(lr_penalty,turned_parameters,cv=5)
grid.fit(X_train,y_train)
print(grid.best_score_)
print(grid.best_params_)
```

## 模型最优化

```python
print(grid.best_params_['penalty'])
print(grid.best_params_['C'])
```

```python
grid.best_estimator_.coef_
```

```python
df=pd.DataFrame({'columns':list(feat_names),'coeffient':list(grid.best_estimator_.coef_.T)})
df.sort_values(by=['coeffient'],ascending=False)
```
